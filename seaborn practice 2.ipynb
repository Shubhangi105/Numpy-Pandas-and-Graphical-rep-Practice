```python
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris,hue='species')
```

```python
g=sns.PairGrid(iris)
g.map(plt.barh)
```

```python
g=sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
g=sns.FacetGrid(data=tips,col='sex',row='smoker')
g.map(sns.distplot,'total_bill')
```

```python
g=sns.FacetGrid(data=tips,col='day',row='sex')
g.map(plt.scatter,'total_bill','tip')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',aspect=0.6,size=8)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.set_context(context='talk', font_scale=0.5)
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='RdYlGn')
```
