```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x=np.linspace(0,5,11)
```

```python
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
plt.xlabel("boo")
plt.ylabel("ooohh")
plt.title("oookkkay")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.4,0.6])
axes.plot(x,y)
axes.set_xlabel("okay")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title("Large")
axes2.plot(y,x)
axes2.set_title("Small")
```

```python
fig, axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
for curr in axes:
    curr.plot(x,y)
#axes.plot(x,y)
```

```python
fig, axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[1].plot(y,x**2)
```

```python
fig=plt.figure(figsize=(4,3))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig.savefig("okay.png",dpi=200)
```

```python
fig=plt.figure(figsize=(4,3))
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,label="okay1")
ax.plot(x**3,x,label="okay2")
ax.legend(loc=0)
```

```python
fig= plt.figure(figsize=(2,2))
ax= fig.add_axes([0,0,2,2])
ax.plot(x,y,color='purple',linewidth=1,linestyle='-',marker='o',markersize=10,markerfacecolor='yellow',markeredgecolor='blue')
```
